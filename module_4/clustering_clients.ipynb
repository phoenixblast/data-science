### Описание признаков
##### Демографические данные

    CustomerID: Уникальный идентификатор клиента.
    BirthYear: Год рождения клиента.
    EducationLevel: Уровень образования клиента.
    MaritalStatus: Семейное положение клиента.
    AnnualIncome: Ежегодный доход домохозяйства клиента.
    NumChildren: Количество детей в семье клиента.
    NumTeenagers: Количество подростков в семье клиента.
    EnrollmentDate: Дата регистрации клиента в компании.

##### Поведение

    DaysSinceLastPurchase: Количество дней с момента последней покупки клиента.
    FiledComplaint: Указано, подавал ли клиент жалобы за последние 2 года (1 — да, 0 — нет).

##### Траты на продукты

    SpentOnWines: Сумма, потраченная на покупку вина за последние 2 года.
    SpentOnFruits: Сумма, потраченная на покупку фруктов за последние 2 года.
    SpentOnMeat: Сумма, потраченная на покупку мяса за последние 2 года.
    SpentOnFish: Сумма, потраченная на покупку рыбы за последние 2 года.
    SpentOnSweets: Сумма, потраченная на покупку сладостей за последние 2 года.
    SpentOnGold: Сумма, потраченная на покупку золота за последние 2 года.

##### Маркетинговые кампании

    DiscountedPurchases: Количество покупок со скидкой.
    AcceptedCampaign1: Указано, принял ли клиент предложение первой кампании (1 — да, 0 — нет).
    AcceptedCampaign2: Указано, принял ли клиент предложение второй кампании (1 — да, 0 — нет).
    AcceptedCampaign3: Указано, принял ли клиент предложение третьей кампании (1 — да, 0 — нет).
    AcceptedCampaign4: Указано, принял ли клиент предложение четвертой кампании (1 — да, 0 — нет).
    AcceptedCampaign5: Указано, принял ли клиент предложение пятой кампании (1 — да, 0 — нет).
    LastCampaignResponse: Указано, принял ли клиент предложение последней кампании (1 — да, 0 — нет).

##### Каналы покупок

    OnlinePurchases: Количество покупок, совершенных через веб-сайт компании.
    CatalogPurchases: Количество покупок, совершенных через каталог.
    InStorePurchases: Количество покупок, совершенных в магазине.
    WebsiteVisitsLastMonth: Количество визитов на веб-сайт компании за последний месяц.

##### Финансовые метрики

    ContactCost: Стоимость контакта с клиентом (постоянная метрика).
    RevenueGenerated: Доход, сгенерированный клиентом (постоянная метрика).

Невалидные значения **Absurd** и **YOLO** составляют малую часть датасета, удаляем

```python
df = df.drop(df[df['MaritalStatus'].isin(['YOLO', 'Absurd'])].index)
df = df.reset_index(drop=True)
```

```python
df['MaritalStatus'] = df['MaritalStatus'].replace({'Alone': 'Single'})
```

Преобразуем временной признак **EnrollmentDate**

```python
df['EnrollmentDate'] = pd.to_datetime(df['EnrollmentDate'], format='%d-%m-%Y')

# Найдем самую раннюю дату регистрации, будем считать ее датой среза данных
last_date = df['EnrollmentDate'].max()

# Создаем новую фичу: вычитаем дату регистрации из последней даты
df['DaysSinceEnrollment'] = (last_date - df['EnrollmentDate']).dt.days

df = df.drop(columns=['EnrollmentDate'])
```

Преобразуем признак **BirthYear** в возраст покупателя для удобства анализа

```python
df['BirthYear'] = last_date.year - df['BirthYear']
df = df.rename(columns={'BirthYear' : 'Age'})
```

Добавим несколько новых признаков. Они лучше переводят данные на язык бизнеса, однако вносят сильную мультиколлинеарность в датасет.

```python
df['TotalSpend'] = df[['SpentOnWines', 'SpentOnFruits', 'SpentOnMeat', 
                        'SpentOnFish', 'SpentOnSweets', 'SpentOnGold']].sum(axis=1)

df['TotalPurchases'] = df['OnlinePurchases'] + df['CatalogPurchases'] + df['InStorePurchases']

df['DiscountRatio'] = df['DiscountedPurchases'] / df['TotalPurchases'].replace(0, 1)

campaign_cols =['AcceptedCampaign1', 'AcceptedCampaign2', 'AcceptedCampaign3', 
                 'AcceptedCampaign4', 'AcceptedCampaign5', 'LastCampaignResponse']
df['TotalPromos_Accepted'] = df[campaign_cols].sum(axis=1)

df['AverageCheck'] = df['TotalSpend'] / df['TotalPurchases'].replace(0, 1)
```

#### Метод K-Means
Найдем оптимальное количество кластеров для алгоритма K-Means используя метод локтя

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


inertia = []
k = range(1, 11)

for k_i in k:
    km = KMeans(n_clusters=k_i, random_state=42)
    km.fit(X)
    inertia.append(km.inertia_)

plt.plot(k, inertia, marker='o', linestyle='-', color='b')
plt.title('График локтя для оптимального количества кластеров')
plt.xlabel('K')
plt.ylabel('Inertia')
plt.grid(True)
```

Несмотря на то, что "локоть" наблюдается уже на 3 кластерах, возьмем больше кластеров, т.к. после 3 инерция все еще ощутимо снижается, и 3 кластера мало для описания выборки клиентов.

Определим **median** и **mean** соответсвенно признакам для оценки распределения кластеров

```python
agg_dict = {}
for col in numeric_features:
    agg_dict[col] = 'median'
for col in binary_features:
    agg_dict[col] = 'mean'
#profiling_columns = df.select_dtypes(include=['number', 'bool']).columns
```

```python
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
km_clusters = kmeans.fit_predict(X)

df.groupby(km_clusters).mean(numeric_only=True).round(2).T
```

```python
print('Распределение кластеров иерархической:')
print(pd.Series(km_clusters).value_counts())
```

Алгоритм KMeans успешно выделил 5 равномерных сегментов с четкими и логичными различиями по уровню дохода, наличию детей и корзине продуктов.

### 3. Профилирование кластеров

#### 0 кластер - Молодые малообеспеченные родители
Малый доход, есть маленькие дети, малый средний чек, много посещают веб-сайт, любят брать товары по скидке.
#### 1 кластер - Взрослые обеспеченные 1
Высокий доход, детей почти нет, сбалансированная корзина, хоть и с перекосом в сторону вина, предпочитают ходить в магазин, слабо откликаются на рекламные предложения.
#### 2 кластер - Взрослые обеспеченные 2
Еще более высокий доход, детей нет, **х2** трат на вино от предыдущего класса, самый высокий отклик на рекламные предложения, но меньше всех выбирают скидки.
#### 3 кластер - Взрослые семьи с детьми
Средний доход, есть дети всех возрастов, малый средний чек, любят скидки, самый низкий отклик на рекламные предложения.
#### 4 кластер - Взрослые семьи с тинейджерами
Доход выше среднего, дети тинейджеры, хороший средний чек, большая часть корзины составляет вино, много покупают через интернет.

```python
# Агрегируем среднее количество покупок по каналам для каждого кластера
channels = df.groupby(km_clusters)[['OnlinePurchases', 'CatalogPurchases', 'InStorePurchases']].mean()

# Строим график с накоплением
channels.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis', edgecolor='black')

plt.title('Предпочтительные каналы покупок по кластерам', fontsize=14)
plt.xlabel('Номер кластера')
plt.ylabel('Среднее количество покупок в год')
plt.legend(['Онлайн (Web)', 'Каталог (Catalog)', 'В магазине (In-Store)'], title='Канал')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import plotly.graph_objects as go

# Выбираем категории трат
categories =['SpentOnWines', 'SpentOnMeat', 'SpentOnFruits', 'SpentOnFish', 'SpentOnSweets', 'SpentOnGold']

# Считаем средние значения для каждого кластера
radar_data = df.groupby(km_clusters)[categories].mean().reset_index()

fig = go.Figure()

colors =['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
cluster_names =[
    '0: Бедные родители', 
    '1: Обеспеченные 1', 
    '2: Обеспеченные 2', 
    '3: Семьи с детьми', 
    '4: Семьи с тинейджерами'
]

# Рисуем многоугольник для каждого кластера
for i, row in radar_data.iterrows():
    fig.add_trace(go.Scatterpolar(
        r=row[categories].values.tolist() + [row[categories].values[0]], 
        theta=categories + [categories[0]],
        fill='toself',
        name=cluster_names[i],
        line_color=colors[i],
        opacity=0.6
    ))

fig.update_layout(
    polar=dict(
        radialaxis=dict(visible=True)
    ),
    title="Профиль продуктовой корзины по кластерам",
    showlegend=True
)

fig.show()
```
