# 14. Какая команда выиграла наибольшое количество матчей в сезоне? (Будьте внимательны с типом матча в сезоне - season).

Сначала посмотрел через **unique** какие сезоны бывают, есть **regular season** и **post season**, видимо имеются в виду только **regular season** игры.

```python
regular_games = data[data['season'] == 'regular season']
home_winners = regular_games[regular_games['home_team_win'] == 1]['home_team']
away_winners = regular_games[regular_games['home_team_loss'] == 1]['away_team']
best_team = pd.concat([home_winners, away_winners]).value_counts()[:1]

print(f'Team "{best_team.index[0]}" won {best_team.values[0]} games')
```

# 15. Какая команда выиграла наибольшее количество домашних матчей в сезоне?

```python
home_team_won = regular_games['home_team_win'] == 1
best_home_team = regular_games[home_team_won]['home_team'].value_counts()[:1]

print(f'{best_home_team.index[0]} won {best_home_team.values[0]} home games')
```

# 16. Какая команда выиграла наибольшее количество гостевых матчей в сезоне?

```python
home_team_lost = regular_games['home_team_loss'] == 1
best_away_team = regular_games[home_team_lost]['away_team'].value_counts()[:1]

print(f'{best_away_team.index[0]} won {best_away_team.values[0]} away games')
```

# 17. Какая команда проиграла наибольшее количество матчей в сезоне?

```python
home_losers = regular_games[regular_games['home_team_win'] == 0]['home_team']
away_losers = regular_games[regular_games['home_team_win'] == 1]['away_team']
worst_team = pd.concat([home_losers, away_losers]).value_counts()[:1]

print(f'{worst_team.index[0]} lost {worst_team.values[0]} games')
```

# 19. Правда ли что большинство проигрышных домашних матчей приходятся на Субботу и Воскресенье?

```python
lost_home_games = data[data['home_team_loss'] == 1]
losses_by_day = lost_home_games['day_of_week'].value_counts(normalize=True)
share_of_Saturday_Sunday = losses_by_day['Saturday'] + losses_by_day['Sunday']

if share_of_Saturday_Sunday > 0.5:
    print(f'Yes, majority ({share_of_Saturday_Sunday:.2%}) of lost home games was played on Saturday and Sunday')
else:
    print(f'No, share of Saturday and Sunday is {share_of_Saturday_Sunday:.2%}')
```

# 20. Правда ли что наибольшее количество ранов происходит в холодную погоду? (Холодной погодой считается погода ниже 0 градусов)

```python
data = pd.read_csv('baseball_games.csv')
data['temperature'] = data.apply(lambda x: round((x['temperature'] - 32 ) / 1.8, 1), axis=1)
number_cold_games = data[data['temperature'] <= 0].index.size
print(f'Total number of games that were played when temperature was 0 or lower: {number_cold_games} of {data.index.size}')
```

Только два матча проводились при холодной погоде, выборка мала чтобы сделать такое утверждение. Тем не менее посмотрим при каких температурах игрались матчи с наибольшим количеством ранов

```python
data.sort_values('total_runs', ascending=False).head(10)[['date', 'temperature', 'total_runs', 'home_team', 'away_team']]
```

Не видно корреляции между температурой и матчами с наибольшим количеством ранов. Можно также посчитать корреляцию для всей выборки

```python
corr_temp_runs = data['total_runs'].corr(data['temperature'])
print(f'Correlation between total runs and temperature is {corr_temp_runs:.2f}')
```

В итоге имеется небольшая корреляция в сторону теплой погоды. Погуглив можно вспомнить что более теплая погода уменьшает сопротивление воздуха, поэтому мяч летит немного дальше.
